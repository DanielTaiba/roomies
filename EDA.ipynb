```python
import os
import pandas as pd
import json

def iter_json_files(path = './generalStats/'):
    for file in os.listdir(path):
        with open(path+file,'r') as f:
            data = json.load(f)
            
            yield data
```

```python
df_countries = pd.DataFrame([data for data in iter_json_files()])
df_countries
```

```python
df_countries.info()
```

```python
df_countries.sort_values(by='roomies',ascending=False).plot(
    kind='bar',
    x='pais',
    figsize=(20,10),
    title='demanda por pais de compartoDepto')
```

```python
df_rooms = pd.DataFrame([data for data in iter_json_files(path='./infoRooms/')])
df_rooms
```
