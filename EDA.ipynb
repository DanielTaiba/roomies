```python
import os
import pandas as pd
import json

def iter_json_files(path = './generalStats/'):
    for file in os.listdir(path):
        with open(path+file,'r') as f:
            data = json.load(f)
            
            yield data
```

```python
df_countries = pd.DataFrame([data for data in iter_json_files()])
df_countries
```

```python
df_countries.info()
```

```python
df_countries.sort_values(by='roomies',ascending=False).plot(
    kind='bar',
    x='pais',
    figsize=(20,10),
    title='demanda por pais de compartoDepto')
```

```python
df_rooms = pd.DataFrame([data for data in iter_json_files(path='./infoRooms/')])
df_rooms
```

```python
df_rooms.info(9)
```

```python
summary_areas = df_rooms.groupby(by=['commune']).agg(mean_cost = ('cost','mean'),count_rooms=('commune','count')).reset_index()
summary_areas
```

```python
summary_areas[['commune','mean_cost']].sort_values(by=['mean_cost'],ascending=False).plot(
    x='commune',
    kind='bar',
    figsize=(20,8),
    title='costo promedio de una pieza por zona en Chile')
```

```python
summary_areas[['commune','count_rooms']].sort_values(by=['count_rooms'],ascending=False).plot(
    x='commune',
    kind='bar',
    figsize=(20,8),
    title='Cantidad de piezas por zona en Chile')
```
